```python
def findArea(r):
    return np.pi*r**2
```

```python
findArea(3)
```

```python
numberList = list(range(0, 10))
numberList
```

```python
for numbers in numberList:
    area = findArea(numbers)
    print("Radius: ", numbers, "Area: ", area)
```
